```python
def sigmoid(x):
    ans = 1 / (1 + np.exp(-x))
    return ans
import numpy as np
import pandas as pd
def logloss(Y, Ycap):
    epsilon = 1e-11
    Y = np.clip(Y, epsilon, 1 - epsilon)
    Ycap = np.clip(Ycap, epsilon, 1 - epsilon)
    Y = np.array(Y)
    Ycap = np.array(Ycap)
    if Y.shape != Ycap.shape:
        raise ValueError("Shapes of Y and Ycap must be the same.")
    Y = np.clip(Y, epsilon, 1 - epsilon)
    ycap = np.clip(Ycap, epsilon, 1 - epsilon)
    first = Y * np.log(ycap)
    second = (1 - Y) * np.log(1 - ycap)
    result = -np.sum(first + second) / len(Y)
    return result
```

```python
# Number of layers
n = int(input("Enter no of layers (excluding output layer): "))
# Incrementing it by 1 for Output layer
n += 1
# Asking for Number of neurons in each layer
L = []
print("\nEnter value except output layer's 1")
for i in range(1, n):
    val = int(input(f"Enter no. of Neuron in Layer {i} :"))
    L.append(val)
L.append(1)
# Initialization of Random weights
W = [0] * n
for i in range(1, n):
    W[i] = np.random.rand(L[i - 1], L[i])
    print("shape",W[i].shape)
Z = [0] * n
A = [0] * n
dz = [0] * n
dw = [0] * n
learning_rate = 0.1
```

```python
def forward_pass():
    for i in range(1, n):
        if i == 1:
            Z[i] = np.dot(X, W[i])
            A[i] = sigmoid(Z[i])
        else:
            Z[i] = np.dot(A[i - 1], W[i])
            A[i] = sigmoid(Z[i])
```

```python
def backward_pass():
    for i in range(n-1, 1, -1):
        if i == n-1:
            dz[i] = A[i] - Y
            dw[i] = np.dot(A[i-1].T, dz[i])
        elif i == 1:
            dz[i] = np.dot(dz[i+1], W[i+1].T) * A[i] * (1 - A[i])
            dw[i] = np.dot(X.T, dz[i])
        else:
            dz[i] = np.dot(dz[i+1], W[i+1].T) * A[i] * (1 - A[i])
            dw[i] = np.dot(A[i-1].T, dz[i])
# Update the weights
    for i in range(n-1, 0, -1):
        W[i] = W[i] - learning_rate * dw[i]
```

```python
def accuracy(Y, Yhat):
    correct_predictions = np.sum(Yhat == Y)
    total_samples = len(Y)
    accuracy = (correct_predictions / total_samples) * 100
    return accuracy
```

```python
for j in range(1, 1000):
    forward_pass()
    # Compute the loss for the current epoch
    loss = logloss(Y, A[n - 1])
    acc = accuracy(Y, (A[n - 1] > 0.5).astype(int)) # Assuming a threshold
    print("\nLoss in epoch :", j, ":", loss)
    print("accuracy", acc)
    backward_pass()
```
